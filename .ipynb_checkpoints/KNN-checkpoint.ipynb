```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Google-Playstore.csv')
df.head()
```

```python
df.sort_values(by = "Rating Count", ascending = False)
```

```python
df.isnull().sum()
```

```python
# Drop records where the app name is missing
df = df[-df['App Name'].isnull()].copy()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df['Currency'].unique()
```

```python
# we will just use USD as default currency
df['Currency'].fillna('USD', inplace = True)
```

```python
df.isnull().sum()
```

```python
# replace missing values for privacy policy with a 0 and for the records that have a privacy policy use value 1

df['Privacy Policy'].fillna(0, inplace = True)
df.loc[df['Privacy Policy'] != 0, 'Privacy Policy'] = 1
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
# get rid of developer website and developer email columns
df.drop('Developer Website', inplace=True, axis=1)
df.drop('Developer Email', inplace=True, axis=1)
```

```python
df.isnull().sum()
```

```python
# get rid of records where the developer id is missing
df = df[-df['Developer Id'].isnull()].copy()
```

```python
df.isnull().sum()
```

```python
# for each record with missing released column, insert the last updated
df['Released'].fillna(df['Last Updated'], inplace = True)

df.isnull().sum()
```

```python
# replace missing minimum android with minimum android mode
minimum_android_mode = df['Minimum Android'].mode()[0]

df['Minimum Android'].fillna(minimum_android_mode, inplace = True)
```

```python
df.isnull().sum()
```

```python
size_mode = df['Size'].mode()[0]

# this will be 'Varies with device'

df['Size'].fillna(size_mode, inplace = True)
```

```python
df.isnull().sum()
```

```python
# replace missing rating with rating mode
# replace missing rating count with 1 (only one rating given...)

rating_mode = df['Rating'].mode()[0]

df['Rating'].fillna(rating_mode, inplace = True)
df['Rating Count'].fillna(1, inplace = True)

df.isnull().sum()
```

```python
# drop all records that have missing number of installs information

df = df[-df['Installs'].isnull()].copy()

df.isnull().sum()
```

```python
# drop scraped time, we don't need this information
df.drop(['Scraped Time'], axis=1, inplace = True)
```

```python
df.head()
```

```python
# converting categorical data into numeric form 
binary_columns = ['Ad Supported', 'In App Purchases', 'Editors Choice', 'Free']

for x in binary_columns:
    df.loc[df[x] == True, x] = 1
    df.loc[df[x] == False, x] = 0
```

```python
df.head(5)
```

```python
df2 = df.copy()
```

```python
drop_columns = ['Developer Id', 'Released', 'Last Updated', 'Currency']

for x in drop_columns:
    df2.drop([x], axis=1, inplace=True)
    
df2.head()
```

```python
df2['Category'].unique().size
```

```python
df_categories = pd.get_dummies(df2['Category'])
df_categories
```

```python
df2 = pd.concat([df2, df_categories], axis=1)
```

```python
df2.head()
```

```python
df_content_rating = pd.get_dummies(df2['Content Rating'])
df_content_rating
```

```python
df2 = pd.concat([df2, df_content_rating], axis=1)
df2.head()
```

```python
# one hot encoding for min android verison
df_min_android = pd.get_dummies(df2['Minimum Android'])
df2['Minimum Android'].unique().size
df2 = pd.concat([df2, df_min_android], axis=1)
df2.head()
```

```python
columns_to_drop = ['App Name', 'App Id', 'Category', 'Installs', 'Minimum Android', 'Content Rating']

# create checkpoint
df3 = df2.copy()


for x in columns_to_drop:
    df3.drop([x], axis=1, inplace=True)
    
df3.head()
```

```python
# want to somehow convert size to numerical format
# 2311541
# CHECKPOINT

df4 = df3.copy()

df4['Size'] = df4['Size'].apply(lambda x: str(x).replace(',', '.') if ',' in str(x) else x)
```

```python
# conversion of kbs into MB
df4['Size'] = df4['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
```

```python
# replace varies with device with 0
df4['Size'] = df4['Size'].apply(lambda x: str(x).replace('Varies with device', '0') if 'Varies with device' in str(x) else x)
```

```python
# convert back to float format
# df4['Size'] = df4['Size'].apply(lambda x: float(x))

df4['Size'] = df4['Size'].apply(lambda x: float(str(x).replace('M', '')) if 'M' in str(x) else x)
df4['Size'] = df4['Size'].apply(lambda x: float(str(x).replace('G', '')) * 1000 if 'G' in str(x) else x)
```

```python
df4['Size'] = df4['Size'].apply(lambda x: float(x))

df4.dtypes['Size']

df4['Size'].head(15)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cos_sim_data = pd.DataFrame(cosine_similarity(df4))

def recommend(index):
    index_recomm =cos_sim_data.loc[index].sort_values(ascending=False).index.tolist()[1:6]
    app_recomm =  df2['App Name'].loc[index_recomm].values
    result = {'Applications':app_recomm,'Index':index_recomm}
    
    return result

# index of instagram ... 
recommend(304824)
    
```
