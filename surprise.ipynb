```python
"""
Simulate a users 5 ratings... (try to do this for like 10 users)

Format should be the following:
user_id, app_id, rating...
"""
import random
import math
# generate the 5 apps that will be rated by the use

def get_random_number(min_score, max_score, probability_power):
    """
    Get a random number between min_score and max_score with a bias towards lower numbers
    That is done because the data is sorted by most popular applications, and the more 
    popular applications have a higher chance of being rated than lower rated ones.
    """
    result = math.floor(min_score + (max_score + 1 - min_score) * (math.pow(random.random(), probability_power)))
    
    return result


df = pd.DataFrame()

for i in range(1,10000):
    possible_number_of_ratings = np.arange(3,20)
    
    ratings = []
    app_ids = []
    user_ids = []
    
    chosen_number_of_ratings = random.choice(possible_number_of_ratings)

    while len(app_ids) != chosen_number_of_ratings:
        num = get_random_number(0, 4999, 10)
        if num not in app_ids:
            app_ids.append(num)
            
    for j in range(0, chosen_number_of_ratings):
        ratings.append(get_random_number(1,5, 0.66))
        
    for x in range(0, chosen_number_of_ratings):
        user_ids.append(i)
        
    data = pd.DataFrame({
            "user_id" : user_ids, 
            "app_id" : app_ids,
            "rating" : ratings
        })
        
    df = pd.concat([df, data])
    
df.head(30)
```

```python
from surprise import KNNWithMeans
from surprise import Dataset
from surprise.model_selection import GridSearchCV

data = Dataset.load_builtin("ml-100k")
sim_options = {
    "name": ["msd", "cosine"],
    "min_support": [3, 4, 5],
    "user_based": [False, True],
}

param_grid = {"sim_options": sim_options}

gs = GridSearchCV(KNNWithMeans, param_grid, measures=["rmse", "mae"], cv=3)
gs.fit(data)

print(gs.best_score["rmse"])
print(gs.best_params["rmse"])
```
