```python
import pandas as pd
import numpy as np
```

```python
df_raw = pd.read_csv('Google-Playstore.csv')
df_raw.head()
```

```python
df_raw.columns
```

```python
df_raw["Rating Count"].describe()
```

```python
df = df_raw.sort_values(by="Rating Count", ascending=False)
```

```python
df.head()
```

```python
# Drop records where the app name is missing
df = df[-df['App Name'].isnull()].copy()
```

```python
df.isnull().sum()
```

```python
df["Currency"].describe()
```

```python
# since USD is the currency of 99% of the dataset, remove all other records that are not USD
df = df[df['Currency'] == 'USD'].copy()
```

```python
# replace missing values for privacy policy with a 0 and for the records that have a privacy policy use value 1

df['Privacy Policy'].fillna(0, inplace = True)
df.loc[df['Privacy Policy'] != 0, 'Privacy Policy'] = 1
```

```python
# do the same for developer website

df['Developer Website'].fillna(0, inplace = True)
df.loc[df['Developer Website'] != 0, 'Developer Website'] = 1
```

```python
# get rid of developer email column
df.drop('Developer Email', inplace=True, axis=1)
```

```python
# get rid of developer Id column
df.drop('Developer Id', inplace=True, axis=1)
```

```python
# for each record with missing released column, insert the last updated
df['Released'].fillna(df['Last Updated'], inplace = True)
```

```python
# replace missing minimum android with minimum android mode
minimum_android_mode = df['Minimum Android'].mode()[0]

df['Minimum Android'].fillna(minimum_android_mode, inplace = True)
```

```python
# replace missing rating with rating mode
# replace missing rating count with 1 (only one rating given...)

rating_mode = df['Rating'].mode()[0]

df['Rating'].fillna(rating_mode, inplace = True)
df['Rating Count'].fillna(1, inplace = True)
```

```python
# drop scraped time, we don't need this information
df.drop(['Scraped Time'], axis=1, inplace = True)
```

```python
# cleaning up the size column
df['Size'] = df['Size'].apply(lambda x: str(x).replace(',', '.') if ',' in str(x) else x)
```

```python
# conversion of kbs into MB
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('k', '')) / 1000 if 'k' in str(x) else x)
```

```python
df['Size'].describe()
```

```python
# replace varies with device with 0
df['Size'] = df['Size'].apply(lambda x: str(x).replace('Varies with device', '19.2') if 'Varies with device' in str(x) else x)
```

```python
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('M', '')) if 'M' in str(x) else x)
df['Size'] = df['Size'].apply(lambda x: float(str(x).replace('G', '')) * 1000 if 'G' in str(x) else x)
```

```python
df2 = df[df['Size'] != str(0)].copy()
```

```python
df2['Size'] = df2['Size'].apply(lambda x: float(str(x)))
```

```python
df2['Size'].describe()
```
